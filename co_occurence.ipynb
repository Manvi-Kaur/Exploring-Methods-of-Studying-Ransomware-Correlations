Choke points points which have mean confidence of occurence > 0.50

```python
ordered_columns = df.mean().sort_values(ascending=False).index
for column in ordered_columns:
    mean_value = df[column].mean()
    print(f"Technique: {column}, Mean: {mean_value}")
```

```python
# Filter columns with mean > 0
filtered_df = df.loc[:, df.mean() > 0]
# filtered_df = filtered_df.sort()
# plt.figure(figsize=(10, 6))
sns.barplot(x=filtered_df.columns, y=filtered_df.mean())
plt.title('Means vs. Techniques ')
plt.xlabel('Techniques')
plt.ylabel('Mean Confidence Score')
plt.xticks(rotation=45)
plt.show()
```

```python
import networkx as nx

techniques = df.columns

# Create a graph
G = nx.Graph()

# Add nodes (techniques) to the graph
G.add_nodes_from(techniques)

# Add edges based on confidence threshold
confidence_threshold = 0.7
for i, source in enumerate(techniques):
    for j, target in enumerate(techniques):
        if i != j and df.iloc[i, j] > confidence_threshold:
            G.add_edge(source, target, weight=df.iloc[i, j])  # Add weight to edges

# Use a circular layout algorithm
pos = nx.circular_layout(G)

# Calculate node sizes based on the number of incoming edges
degrees = dict(G.degree())
max_degree = max(degrees.values())
min_degree = min(degrees.values())
node_sizes = [1000 + 3000 * (degrees[node] - min_degree) / (max_degree - min_degree) for node in G.nodes()]
plt.figure(figsize=(10, 10))
# Draw all edges in gray
edges = G.edges()
nx.draw_networkx_edges(G, pos, edgelist=edges, width=1.0, alpha=0.5, edge_color='gray')

# Draw nodes with fixed color
nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color='skyblue')

# Draw labels
labels = {node: node for node in G.nodes()}
nx.draw_networkx_labels(G, pos, labels=labels, font_size=10)

plt.title('Techniques Network')
plt.axis('off')  # Disable axis
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python

# Create a graph
G = nx.DiGraph()

# Add nodes (techniques) to the graph
G.add_nodes_from(techniques)

# Add edges based on confidence threshold
confidence_threshold = 0.7
for i, source in enumerate(techniques):
    for j, target in enumerate(techniques):
        if i != j and df.iloc[i, j] > confidence_threshold:
            G.add_edge(source, target, weight=df.iloc[i, j])  # Add weight to edges

# Use a circular layout algorithm
pos = nx.circular_layout(G)
plt.figure(figsize=(10, 10))
# Calculate node sizes based on the number of edges connected
degrees = dict(G.degree())
max_degree = max(degrees.values())
min_degree = min(degrees.values())
node_sizes = [1000 + 3000 * (degrees[node] - min_degree) / (max_degree - min_degree) for node in G.nodes()]

# Calculate node colors based on the number of edges connected
node_colors = [(degree - min_degree) / (max_degree - min_degree) for degree in degrees.values()]

# Draw all edges in gray
edges = G.edges()
nx.draw_networkx_edges(G, pos, edgelist=edges, width=1.0, alpha=0.5, edge_color='gray', arrows = True)

# Draw nodes with gradient colors based on the number of edges connected
nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color=node_colors, cmap=plt.cm.Blues, vmin=-0.5, vmax=1.2)

# Draw labels
labels = {node: node for node in G.nodes()}
nx.draw_networkx_labels(G, pos, labels=labels, font_size=10)

plt.title('Techniques Network')
plt.axis('off')  # Disable axis
plt.tight_layout()  # Adjust layout to prevent overlap

plt.show()
```

```python
tactic_df_reset = tactic_df.reset_index()
```

```python
from matplotlib.patches import Patch

# Assuming df is the adjacency matrix and tactic_df contains tactic information

# Step 1: Group techniques based on their associated tactics
grouped_techniques = tactic_df_reset.groupby('TacticID')['TechniqueID'].apply(list)

# Step 2: Create a directed graph
G = nx.DiGraph()

# Step 3: Add nodes to the graph with their tactic information
for tactic, techniques in grouped_techniques.items():
    for technique in techniques:
        G.add_node(technique, tactic=tactic)

# Step 4: Add edges to the graph with confidence scores greater than 80 and ensure only from lower to higher tactic
for i, row in df.iterrows():
    for j, col in row.items():  # Change from iteritems to items
        if col > 0.7 and tactic_df.loc[i, 'TacticID'] < tactic_df.loc[j, 'TacticID']:
            G.add_edge(i, j, weight=col)

# Visualize the graph
plt.figure(figsize=(10, 12))

# Group nodes based on tactic and set node color accordingly
colors = {'TA0001': sns.color_palette("Blues", 11)[3],  # Light blue
          'TA0002': sns.color_palette("Greens", 11)[3],  # Light green
          # 'TA0003': sns.color_palette("Blues", 11)[7],  # Medium blue
          'TA0004': sns.color_palette("Greens", 11)[7],  # Medium green
          'TA0005': sns.color_palette("Blues", 11)[5],  # Dark blue
          'TA0006': sns.color_palette("Greens", 11)[9],  # Dark green
          'TA0007': sns.color_palette("Blues", 11)[2],  # Extra dark blue
          # 'TA0008': sns.color_palette("Greens", 11)[2],  # Extra dark green
          'TA0009': sns.color_palette("Blues", 11)[10],  # Navy blue
          'TA0010': sns.color_palette("Greens", 11)[8],  # Forest green
          'TA0011': sns.color_palette("Blues", 11)[8]}  # Navy blue

groups = sorted(set(nx.get_node_attributes(G, 'tactic').values()))
pos = {}
y_offset = 0
for group in groups:
    nodes = [node for node, attr in G.nodes(data=True) if attr['tactic'] == group]
    for i, node in enumerate(nodes):
        pos[node] = (i, y_offset)
    y_offset -= 1

# Draw nodes with specified positions
node_legend_handles = []
for tactic, color in colors.items():
    dummy_node = Patch(color=color, label=tactic)
    node_legend_handles.append(dummy_node)
    plt.scatter([], [], color=color, label=tactic)  # Dummy scatter plot to add nodes to legend

nx.draw_networkx_nodes(G, pos, node_size=1000, node_color=[colors[G.nodes[node]['tactic']] for node in G.nodes()])
nx.draw_networkx_labels(G, pos, labels={node: node for node in G.nodes()}, font_size=8)

# Draw directed edges with labels
edges = G.edges()
edge_labels = {(u, v): f"{G[u][v]['weight']:.2f}" for u, v in edges}
nx.draw_networkx_edges(G, pos, arrows=True, edge_color='gray', arrowsize=20)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8)

plt.title('Attack Scenario Visualization')
plt.legend(handles=node_legend_handles, title='TacticID', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.axis('off')
plt.show()
```
